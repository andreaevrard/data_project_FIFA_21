```python
def transform_money(c):
    c = c[1:]
    if c[-1] == 'K':
        res = float(c[:-1]) * 1000
    elif c[-1] == 'M':
        res = float(c[:-1]) * 1000000
    else:
        res = float(c)
    return res
```
